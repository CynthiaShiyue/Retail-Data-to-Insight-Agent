# Construct a sample dataset

```python
import pandas as pd

# https://www.kaggle.com/datasets/manjeetsingh/retaildataset

features = pd.read_csv("Features data set.csv")
sales = pd.read_csv("sales data-set.csv")

features = features[["Store", "Date", "Temperature", "Fuel_Price", "CPI", "Unemployment", "IsHoliday"]]

features.columns = features.columns.str.lower()
sales.columns = sales.columns.str.lower()

for df in [features, sales]:
    df["store"] = df["store"].astype(str)
    df["date"] = pd.to_datetime(df["date"], format="%d/%m/%Y")

merged = pd.merge(sales, features, on=["store", "date"], how="left")

merged.to_csv("(sample2)retaildata.csv", index=False)


```
